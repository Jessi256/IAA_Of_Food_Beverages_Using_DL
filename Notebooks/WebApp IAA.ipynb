```python
#5. Run this file to create app.py
%%writefile app.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import PIL
import PIL.Image as img
import cv2
import os
import tensorflow as tf
import keras.utils as ker_image
from tensorflow.keras.utils import img_to_array
st.set_option('deprecation.showfileUploaderEncoding', False)

##################################################################################

model = tf.keras.models.load_model('/content/drive/MyDrive/Model/InceptionV3_model.h5') #Or load in another model from DataScience Sharepoint
size=299

def load_image(img_path, size):

    IAA_image = ker_image.load_img(file, target_size=(size,size))
    img_tensor = ker_image.img_to_array(IAA_image)                    # (height, width, channels)
    img_tensor = np.expand_dims(img_tensor, axis=0)         # (1, height, width, channels), add a dimension because the model expects this shape: (batch_size, height, width, channels)
    img_tensor /= 255.                                      # imshow expects values in the range [0, 1]

    return img_tensor

##################################################################################
#https://www.youtube.com/watch?v=dw96S_iFFbI&ab_channel=DiazonicLabs


st.title('Image Aesthetics Assessment \n ## of Food and Beverages')

file= st.file_uploader("Choose a food or beverage image to predict the aesthetics...", type=["png", "jpg", "jpeg"])

if file is not None:
  image = img.open(file)
  st.image(
  image,
  caption=f"Uploaded Image",
  use_column_width=True)

  if st.button ("Predict"):
    img_tensor= load_image(file,size)
    pred = model.predict(img_tensor)
    pred_value_round=[ '%.4f' % elem for elem in pred[0]]
    #pred_val=float(pred_value_round[0]) *100

    if float(pred_value_round[0]) >= 0.5:
        pred_percentage= float(pred_value_round[0]) *100
        output_text= f'Your image is {str(pred_percentage)[:5]} % aesthetic. I am hungry now!'
    else:
        pred_percentage=100 - (float(pred_value_round[0]) *100)
        output_text= f'Your image is {str(pred_percentage)[:5]} % unaesthetic. I would not eat that if I were you...  :wink:'

    st.info(f' ### :trophy: {output_text}')    
   
```
